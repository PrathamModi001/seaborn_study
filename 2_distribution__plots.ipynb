```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
info = pd.read_csv("StudentsPerformance.csv")
df = pd.read_csv("dm_office_sales.csv")
df.head()
```

```python
# sns.set(style="darkgrid") # Adds a grid
sns.histplot(x="salary", data=df, bins=30,kde=True); # 200 vertical lines
# kde = Kernel Distribution Estimation: Probability Curve Jaisa
```

# Statistics
To show some kind of stats through visualisation

## Countplot

```python
df["division"].value_counts()
```

```python
plt.figure(figsize=(10,3), dpi=200)
sns.countplot(x="division", data=df, hue="level of education")
```

## Barplot

This is saying:
x mein level of education ko rakho,
for each category of x, unki salary batao ==>
BY DOING SOME FUCNTION on the salaries of categories

SO we are showing:

np.mean(Average) Salary per level of education per division if we also add hue = division

```python
plt.figure(figsize=(10,4), dpi=200)
sns.barplot(data=df,x="level of education", y="salary", estimator=np.mean, errorbar='sd', hue="division")

plt.tight_layout()
plt.legend(loc=(1,0.3))
```
