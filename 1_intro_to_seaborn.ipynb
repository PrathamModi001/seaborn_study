```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
df = pd.read_csv("dm_office_sales.csv")
df.head()
```

Note that we can call the figure here and change its parameters as well

```python
plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x="salary", y="sales",data=df, s=150, alpha=0.2);

# s=200 increases the size of all markers
# Since multiple markers are inevitably going to stack over one another we can change the opacity of the points
```

# GETTING 3D INFORMATION
3 or more information from the plot

## Hue = Some Categorical Feature Column
(like color, division)

```python
plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x="salary", y="sales",data=df, hue="level of education");
```

## Style:
Note that it should be a categorical feature

```python
plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x="salary", y="sales",data=df, hue="level of education" ,style='division');
```

## Hue = Continuous Column
 (like salary, age)

```python
plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x="salary", y="sales",data=df, hue="salary");
```

### Colormaps in Matplotlib
https://matplotlib.org/stable/tutorials/colors/colormaps.html

```python
plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x="salary", y="sales",data=df, hue="salary", palette="plasma");
```

## Size

```python
plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x="salary", y="sales",data=df, hue="salary", size="work experience");
```
